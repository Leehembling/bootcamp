Uses .loc to return 'name' column, all rows

**Loc and iloc are two functions in Pandas that are used to slice a data set in a Pandas DataFrame. The function .loc is typically used for label indexing and can access multiple columns, while .iloc is used for integer indexing.  **

Source: https://builtin.com/software-engineering-perspectives/pandas-iloc

```python
column_data = df.loc[:, 'name']  # Assuming 'name' is a valid column name
print(column_data)
```

```python
# Script 1: Filter rows where the 'mark' is greater than 70
filtered_rows = df[df['mark'] > 70]
print(filtered_rows)

# Note: This script selects rows from the DataFrame `df` where the values in the 'mark' column are greater than 70.
# It uses boolean indexing to apply the condition to the DataFrame, resulting in a filtered DataFrame where only rows
# that meet the condition are included. The resulting DataFrame is stored in `filtered_rows`, which is then printed.
```

```python
# Script 2: Filter rows where the 'mark' is greater than 70 AND less than 85
filtered_rows_and = df[(df['mark'] > 70) & (df['mark'] < 85)]
print(filtered_rows_and)

# Note: In this script, two conditions are applied to the 'mark' column of the DataFrame `df`:
# 1. The 'mark' should be greater than 70.
# 2. The 'mark' should be less than 85.
# These conditions are combined using the logical AND operator (&), meaning a row must satisfy both conditions
# to be included in the filtered result. The filtered DataFrame, `filtered_rows_and`, contains only the rows that
# meet both criteria and is printed to show the result.
```

```python
# Script 1: Filter rows where 'mark' is greater than 50 OR less than 20
filtered_rows_or = df[(df['mark'] > 50) | (df['mark'] < 20)]
print(filtered_rows_or)
```
